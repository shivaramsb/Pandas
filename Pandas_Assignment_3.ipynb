**1.Select rows by label using .loc.**

```python
import pandas as pd
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
df.loc[0]
```

**2. Select rows by position using .iloc.**

```python
df.iloc[0]
```

**3. Set a column as the index.**

```python
df.set_index('A')
```

**4. Reset the index back to default.**

```python
df.reset_index()
```

**5. Select rows with specific conditions using .loc.**

```python
df.loc[df['A'] > 2]
```

**6. Select a range of rows using .iloc.**

```python
df.iloc[0:2]
```

**7. Select specific rows and columns using .loc.**

```python
df.loc[0, 'A']
```

**8. Select specific rows and columns using .iloc.**

```python
df.iloc[0, 0]
```

**9. Conditional selection using multiple conditions.**

```python
df[(df['A'] > 1) & (df['B'] < 6)]
```

**10. Use the .query() method to filter rows.**

```python
df.query('A > 1 and B < 6')
```
