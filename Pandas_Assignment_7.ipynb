```python
import pandas as pd
import numpy as np

# Generate date range
date_range = pd.date_range(start='2023-01-01', end='2023-01-31')

# Create a DataFrame with random stock prices for three companies
np.random.seed(42)
data = {
    'date': date_range.tolist() * 3,
    'company': ['CompanyA']*31 + ['CompanyB']*31 + ['CompanyC']*31,
    'price': np.random.normal(loc=100, scale=10, size=31).tolist() +
             np.random.normal(loc=200, scale=20, size=31).tolist() +
             np.random.normal(loc=300, scale=30, size=31).tolist()
}

df = pd.DataFrame(data)
print(df)
```

**1.Plot a line chart.**

```python
df.plot.line()

```

**2. Plot a bar chart.**

```python
df.plot.bar()
```

**3. Plot a histogram.**

```python
df.plot.hist()
```

**4. Plot a box plot.**

```python
df.plot.box()
```

**5. Plot a scatter plot.**

```python
df.plot.scatter(x='date', y='price')
```

**6. Plot a pie chart.**

```python
df.plot.pie(y='price')
```

**7. Customize the plot title and labels.**

```python
df.plot(title='My Plot', xlabel='X-axis', ylabel='Y-axis')

```

**8. Change the plot style.**

```python
pd.set_option('plotting.backend', 'matplotlib')
df.plot(style='--')
```

**9. Save a plot to a file.**

```python
fig = df.plot().get_figure()
fig.savefig('plot.png')
```

**10. Plot multiple plots in a single figure.**

```python
df.plot(subplots=True)

```
