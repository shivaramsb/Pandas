**1.Handle missing data by dropping rows.**

```python
import pandas as pd
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
df.dropna()
```

**2. Handle missing data by filling with a specific value.**

```python
df.fillna(0)
```

**3. Identify duplicated rows.**

```python
df.duplicated()
```

**4. Remove duplicated rows.**

```python
df['A'].replace(1, 100)
```

**5. Replace values in a column.**

```python
df['A'].replace(1, 100)
```

**6. Change data types of a column.**

```python
df['A'] = df['A'].astype(float)
```

**7. Rename the index of a DataFrame.**

```python
df.rename_axis('new_index')
```

**8. Handle missing data by forward fill.**

```python
df.fillna(method='ffill')
```

**9. Handle missing data by backward fill.**

```python
df.fillna(method='bfill')
```

**10. Standardize column names to lower case.**

```python
df.columns = [col.lower() for col in df.columns]
```
