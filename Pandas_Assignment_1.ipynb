```python
df.describe()
```

```python
df['A']
```

**5. Select multiple columns from a DataFrame.**

```python
df[['A', 'B']]
```

```python
df[df['A'] > 1]
```

```python
df.sort_values(by='A')
```

```python
df.reset_index(drop=True)
```
